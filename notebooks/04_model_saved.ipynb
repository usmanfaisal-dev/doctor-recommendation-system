```python
import pandas as pd
import os
import pickle

DATA_DIR = "D:/Doc_Recommendation/data"
MODEL_DIR = "D:/Doc_Recommendation/model"
os.makedirs(MODEL_DIR, exist_ok=True)

# Load doctors CSV
doctors_df = pd.read_csv(os.path.join(DATA_DIR, "doctors_cleaned.csv"))

# Convert percentage strings to numeric
if doctors_df['success_rate'].dtype == object:
    doctors_df['success_rate'] = doctors_df['success_rate'].str.replace('%', '').astype(float) / 100

# Ensure specialty_code exists
if "specialty_code" not in doctors_df.columns:
    doctors_df["specialty_code"] = doctors_df["specialty"].astype("category").cat.codes

# Save pickle
doctors_df.to_pickle(os.path.join(MODEL_DIR, "doctors_data.pkl"))

print("doctors_data.pkl saved with numeric success_rate!")
```
